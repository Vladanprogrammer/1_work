```python
class PalindromeChecker:
    """Клас, який приймає рядок, обробляє його та перевіряє на паліндром."""
    def __init__(self, text, student_name='Не вказано', group='Не вказано'):
        self.text = text
        self.student_name = student_name
        self.group = group
        self._normalized = None
        self._is_palindrome = None

    def process_data(self):
        """Нормалізує текст (видаляє небуквено-цифрові символи, пробіли, приводить до нижнього регістру)."""
        s = str(self.text).lower()
        self._normalized = ''.join(ch for ch in s if ch.isalnum())
        self._is_palindrome = (self._normalized == self._normalized[::-1])
        return self._is_palindrome

    @property
    def is_palindrome(self):
        if self._is_palindrome is None:
            return self.process_data()
        return self._is_palindrome

    @property
    def result(self):
        return f"Is palindrome: {self.is_palindrome}"

    def __str__(self):
        return f"Студент: {self.student_name}, Група: {self.group}"

samples = [
    ('Level', 'Іван Іванов', 'CS-01'),
    ('А роза упала на лапу Азора', 'Марія Петрівна', 'CS-02'),
    ('Hello, world!', 'Олег Студент', 'CS-03'),
]
for text, name, group in samples:
    pc = PalindromeChecker(text, student_name=name, group=group)
    print(pc)
    pc.process_data()
    print('Input:', repr(pc.text))
    print(pc.result)
    print('Normalized:', pc._normalized)
    print('-' * 40)
```

## Створити клас PalindromeChecker, який:

### Приймає в конструкторі рядок
### Має метод check(), який перевіряє чи є рядок паліндромом (читається однаково в обох напрямках)
### Має property is_palindrome, яка повертає True або False

```python
class PalindromeChecker:
    """Перевіряє, чи є рядок паліндромом. Має метод `check()` і property `is_palindrome`."""
    def __init__(self, text):
        self.text = text
        self._is_palindrome = None

    def check(self):
        """Нормалізує рядок (залишає лише буквено-цифрові символи, приводить до нижнього регістру) і перевіряє паліндром."""
        s = ''.join(ch.lower() for ch in str(self.text) if ch.isalnum())
        self._is_palindrome = (s == s[::-1])
        return self._is_palindrome

    @property
    def is_palindrome(self):
        if self._is_palindrome is None:
            return self.check()
        return self._is_palindrome

# Приклади: мінімум 2-3 об'єкти із різними вхідними даними (вивід українською)
examples = ['Level', 'Hello', 'А роза упала на лапу Азора']
for ex in examples:
    pc = PalindromeChecker(ex)
    print(f"Рядок: {ex!r}")
    print(f"  Результат перевірки (check()): {pc.check()}")
    print(f"  Паліндром: {pc.is_palindrome}")
    print('-' * 40)
```
