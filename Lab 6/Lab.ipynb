1 Завдання: Інкапсуляція 
Додайте генератор випадкових чисел та викличіть методи deposit та withdraw у циклі передаючи випадкові числа. Виведіть кінцевий результат;

```python
import random  

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  # публічний атрибут
        self.__balance = balance  # приватний атрибут

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.__balance

account = BankAccount("Bohdan", 1000)
print(f"Початковий баланс клієнта {account.owner}: {account.get_balance()} грн")
print("-" * 30)

for i in range(5):
    rand_deposit_amount = random.randint(50, 800)
    rand_withdraw_amount = random.randint(50, 800)
    
    print(f"Крок {i + 1}:")
    
    account.deposit(rand_deposit_amount)
    print(f"  [+] Поповнено на: {rand_deposit_amount} грн. Поточний баланс: {account.get_balance()} грн.")
    
    withdraw_status = account.withdraw(rand_withdraw_amount)
    
    if withdraw_status == "Insufficient funds":
        print(f"  [-] Спроба зняти: {rand_withdraw_amount} грн. Відмовлено: {withdraw_status}!")
    else:
        print(f"  [-] Знято: {rand_withdraw_amount} грн. Поточний баланс: {account.get_balance()} грн.")
        
    print("-" * 30)

print(f"Кінцевий результат (Баланс): {account.get_balance()} грн.")
```

2 Завдання Наслідування
Створіть ще один метод у класі Vehicle та викличіть його з обєкта класу Car;

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"
    
    def start_price(self):
        return 10000  

class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)  # виклик конструктора базового класу
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Price: ${self.start_price()}, Seats: {self.seats}"

car = Car("Toyota", "Camry", 5)
print(car.display_info())
```

3 Завдання Поліморфізм
Створіть клас Fish у якого небуде метода speak. Як буде поводити себе обєкт класу Fish при виклику методу speak?

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"
    
class Fish(Animal):
    pass   

# Використання:
animals = [Dog(), Cat(), Fish()]
for animal in animals:
    print(animal.speak())
```

4 Завдання Абстакція
Дописуємо гру Додайте третій тип зброї — Bow (лук), який:

    має власний параметр range_power (дальність);
    метод attack() з формулою шкоди attack_power + randint(5, 15) + range_power;
    має метод reload(), який збільшує дальність (range_power += 1);

Додайте випадковий вибір зброї з трьох можливих (Sword, Axe, Bow);

Зробіть гру покроковою, де користувач може робити хід та вибирати дії з своїм типом зброї (наприклад, атакувати або накласти підсилення щоб підвищити ефективність зброї на наступному кроці);

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name: str, health=500):
        self.name = name
        self.health = health
    
    @abstractmethod
    def attack(self, another_item):
        pass

class Sword(Item): # Це наслідування
    def __init__(self, name: str, attack_power: int):
        super().__init__(name=name) #  Це виклик конструктора наслідуваного класу
        self.__attack_power = attack_power # приватний атрибут який реалізує інкапсуляцію
        self.__sharp = 0
    
    def attack(self, another_item: Item): # ми не можемо створити клас без цього методу
        current_attack = self.__attack_power + self.__sharp + randint(0, 10)
        another_item.health -= current_attack
        return f"Меч {self.name} наносить {current_attack} шкоди. Здоров'я {another_item.name}: {another_item.health}"
    
    def sharpening(self):
        self.__sharp += 10
        return f"Меч {self.name} нагострено! Сила удару зросла."

class Axe(Item): # Це наслідування
    def __init__(self, name: str, attack_power: int):
        super().__init__(name=name) #  Це виклик конструктора наслідуваного класу
        self.__attack_power = attack_power # приватний атрибут який реалізує інкапсуляцію
        self.__rage = 0
    
    def attack(self, another_item: Item):  # ми не можемо створити клас без цього методу
        current_attack = self.__attack_power + self.__rage + randint(0, 20)
        another_item.health -= current_attack
        self.__rage = 0 
        return f"Сокира {self.name} наносить {current_attack} шкоди. Здоров'я {another_item.name}: {another_item.health}"

    def enrage(self):
        self.__rage += 15
        return f"Сокира {self.name} входить у стан люті! Наступний удар буде нищівним."

class Bow(Item):
    def __init__(self, name: str, attack_power: int, range_power: int = 5):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self.__range_power = range_power
        self.__numb_arrows = 6

    def attack(self, another_item: Item):
        if self.__numb_arrows <= 0:
            return f"У {self.name} закінчилися стріли! Потрібно перезарядити."
        current_attack = self.__attack_power + randint(5, 15) + self.__range_power
        another_item.health -= current_attack
        self.__numb_arrows -= 1
        return f"Лук {self.name} наносить {current_attack} шкоди. Здоров'я {another_item.name}: {another_item.health}. Стріл: {self.__numb_arrows}"
    
    def reload(self):
        self.__range_power += 2
        self.__numb_arrows = 6
        return f"Лук {self.name} перезаряджено. Дальність стрільби збільшено до {self.__range_power}!"


available_weapons = [
    Sword("Ескалібур", 30),
    Axe("Кратос", 35),
    Bow("Соколине Око", 25)
]

player_weapon = choice(available_weapons)
enemy = Axe("Орк-Бос", 40)

print(f"\nГра почалась! Ваша випадкова зброя: {player_weapon.__class__.__name__} '{player_weapon.name}'")
print(f"Ваш ворог: {enemy.name} ({enemy.health} HP)\n")

for i in range(1, 11):
    print(f"\n--- Хід {i} ---")
    print("Оберіть дію:")
    print("1. Атакувати ворога")
    print("2. Використати підсилення зброї (Нагострити/Лють/Перезарядити)")
    
    action = input("Ваш вибір (1 або 2): ")
    
    if action == "1":
        print(player_weapon.attack(enemy))
    elif action == "2":
        if isinstance(player_weapon, Sword):
            print(player_weapon.sharpening())
        elif isinstance(player_weapon, Axe):
            print(player_weapon.enrage())
        elif isinstance(player_weapon, Bow):
            print(player_weapon.reload())
    else:
        print("Неправильний вибір! Ви пропускаєте хід.")
        
    if enemy.health <= 0:
        print(f"\nВітаємо! Ви перемогли {enemy.name}!")
        break
        
    print(f"\nХід ворога...")
    print(enemy.attack(player_weapon))
    
    if player_weapon.health <= 0:
        print(f"\nВас знищено. Перемога за {enemy.name}!")
        break
```
