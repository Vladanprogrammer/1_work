```python
from enum import Enum
import random


class BookState(Enum):
    AVAILABLE = "доступна"
    BORROWED = "позичена"


class Book:
    def __init__(self, title: str):
        self.title = title
        self.state = BookState.AVAILABLE

    def __repr__(self):
        return f"Book(title='{self.title}', state='{self.state.value}')"
```
